```python
#Student marks management system
#Making list of students
student_marks = [85, 92, 78, 65, 90, 72]

print("Student Marks:", student_marks)
#Display highest and lowest marks
highest_mark = max(student_marks)
lowest_mark = min(student_marks)

print(f"Highest mark: {highest_mark}")
print(f"Lowest mark: {lowest_mark}")
#Calcualte average marks
total_marks = sum(student_marks)
number_of_students = len(student_marks)

# Check to prevent division by zero if the list is empty
if number_of_students > 0:
    average_marks = total_marks / number_of_students
    print(f"Average marks: {average_marks:.2f}")
else:
    print("Cannot calculate average: The list of student marks is empty.")
#Sorting the marks
sorted_marks = sorted(student_marks)

print(f"Sorted marks: {sorted_marks}")
```
